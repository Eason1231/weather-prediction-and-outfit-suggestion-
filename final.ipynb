```python
def clothes_male(maxtemp,mintemp):
# 準備訓練資料
    X = [[35],[30], [25], [20], [15], [10]]  # 溫度
    y = ['無袖','短袖', '薄長袖或短袖加薄外套', '長袖加薄外套', '厚長袖加厚外套', '多穿幾件厚衣服加厚外套']  # 穿著

# 建立決策樹模型
    model = tree.DecisionTreeClassifier()

# 訓練模型
    model.fit(X, y)
    
# 預測新的溫度
    new_temperature1 = [[maxtemp]]  # 要預測的溫度
    predicted_attire1 = model.predict(new_temperature1)
    new_temperature2 = [[mintemp]]  # 要預測的溫度
    predicted_attire2 = model.predict(new_temperature2)
# 輸出預測結果
    return predicted_attire1,predicted_attire2
```

```python
def clothes_female(maxtemp,mintemp):
# 準備訓練資料
    X = [[35],[30], [25], [20], [15], [10]]  # 溫度
    y = ['無袖或運動內衣','短袖、短袖洋裝', '薄長袖或長袖洋裝加薄外套', '長袖薄外套', '厚長袖加厚外套', '多穿幾件厚衣服加厚外套']  
    # 穿著

# 建立決策樹模型
    model = tree.DecisionTreeClassifier()

# 訓練模型
    model.fit(X, y)
    
# 預測新的溫度
    new_temperature1 = [[maxtemp]]  # 要預測的溫度
    predicted_attire1 = model.predict(new_temperature1)
    new_temperature2 = [[mintemp]]  # 要預測的溫度
    predicted_attire2 = model.predict(new_temperature2)
# 輸出預測結果
    return predicted_attire1,predicted_attire2
```

```python
def pants_female(maxtemp,mintemp):
# 準備訓練資料
    X = [[35],[30], [25], [20], [15], [10]]  # 溫度
    y = ['短褲或短裙','短褲或短裙', '長裙或瑜珈褲', '薄長褲', '厚長褲', '厚長褲或多層褲']  # 穿著

# 建立決策樹模型
    model = tree.DecisionTreeClassifier()

# 訓練模型
    model.fit(X, y)
    
# 預測新的溫度
    new_temperature1 = [[maxtemp]]  # 要預測的溫度
    predicted_attire1 = model.predict(new_temperature1)
    new_temperature2 = [[mintemp]]  # 要預測的溫度
    predicted_attire2 = model.predict(new_temperature2)
# 輸出預測結果
    return predicted_attire1,predicted_attire2
```

```python
def rain(rainchance):
# 準備訓練資料
    X = [[70],[50], [30]]  # 降雨機率
    y = ['攜帶雨具','可以準備輕便雨衣或小摺疊傘', '不需要擔心下雨喔:)']  # 穿著

# 建立決策樹模型
    model = tree.DecisionTreeClassifier()

# 訓練模型
    model.fit(X, y)
    
# 預測新的降雨機率
    new_rain = [[rainchance]]  # 要預測的降雨機率
    predicted_attire = model.predict(new_rain)
    
# 輸出預測結果
    return predicted_attire
```

```python
def uv(uv):
# 準備訓練資料
    X = [[10], [7],[5],[2]]  # uv
    y = ['使用SPF 30以上之防曬產品，且每兩小時重新塗抹'\
         ,'使用至少含有SPF 30的防曬產品，並定期重新塗抹', '使用含有SPF 30的防曬產品，並注意防曬','使用含有SPF 15的防曬產品']  
# 建立決策樹模型
    model = tree.DecisionTreeClassifier()

# 訓練模型
    model.fit(X, y)
    
# 預測新的uv
   
    if(uv=='-1'):
        predicted_attire ='此地無觀測站，請查看其他地區'
    else:
        new_uv = [[uv]]  # 要預測的uv
        predicted_attire = model.predict(new_uv)   
    
# 輸出預測結果
    return predicted_attire
```
